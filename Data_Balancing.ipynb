```python
import os

# Folder where your .wav files are stored
input_folder = "Neutral"
output_folder = "Relabled"  

# Fixed RAVDESS parameters
modality = "03"
vocal_channel = "01"
emotion = "01"      # Neutral
repetition = "01"
statement = "03"    # depending on which sentence was spoken
gender = "01"

# Start actor ID from 01
start_actor_id = 1

# Loop over all WAV files and rename
for idx, filename in enumerate(sorted(os.listdir(input_folder))):
    if filename.lower().endswith(".wav"):
        actor_id = start_actor_id + idx
        actor_str = f"{actor_id:02d}"

        
        # Create new filename in RAVDESS format
        new_filename = f"{modality}-{vocal_channel}-{emotion}-{statement}-{repetition}-{actor_str}-{gender}.wav"

        # Full paths
        old_path = os.path.join(input_folder, filename)
        new_path = os.path.join(output_folder, new_filename)

        # Rename the file
        os.rename(old_path, new_path)
        print(f"Renamed: {filename} → {new_filename}")
```
